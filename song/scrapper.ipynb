```python
df1 = pd.read_csv("emea_apac_all_players_stats.csv")
```

```python
df2 = pd.read_csv("emea.csv")
```

```python
df3 = pd.read_csv("apac.csv")
```

```python
# calculate KPR for each player-agent combination in df1
df1["Player-Agent KPR"] = (df1["Kills"] / df1["Rounds Played"]).round(3)
df1
```

```python
# calculate KPR for each agent-map combination in df2 and df3 using MapScore
def calculate_rounds_played(map_score):
    try:
        scores = map_score.split("‑")
        return int(scores[0]) + int(scores[1])
    except:
        return None
```

```python
# aggregate data at the player-agent-map level for df2 and df3
df2_agg = df2.groupby(['Player', 'Agent', 'Map']).agg(
    Kills=('Kills', 'sum'),
    Rounds_Played=('MapScore', lambda x: sum(calculate_rounds_played(score) for score in x))
).reset_index()


df2_agg['Player-Agent-Map KPR'] = (df2_agg['Kills'] / df2_agg['Rounds_Played']).round(3)

df3_agg = df3.groupby(['Player', 'Agent', 'Map']).agg(
    Kills=('Kills', 'sum'),
    Rounds_Played=('MapScore', lambda x: sum(calculate_rounds_played(score) for score in x))
).reset_index()

df3_agg['Player-Agent-Map KPR'] = (df3_agg['Kills'] / df3_agg['Rounds_Played']).round(3)
```

```python
# Combine aggregated data from df2 and df3
combined_agg = pd.concat([df2_agg, df3_agg], ignore_index=True)
combined_agg
```

```python
# merge df1 and combined_agg by aligning the column names
df = pd.merge(df1, combined_agg, left_on=['Name', 'Agent'], right_on=['Player', 'Agent'], how='inner')

# drop the redundant 'Player' column after the merge
df= df.drop(columns=['Player'])
```

```python
# compute the mean KPR
df['Mean KPR'] = df[['Player-Agent KPR', 'Player-Agent-Map KPR']].mean(axis=1)

# display the resulting dataframe
df
```

```python
def predict_agent(player, map_name, df):
    """
    Predict the most likely agent a player will play on a given map.

    Parameters:
    - player (str): The name of the player.
    - map_name (str): The name of the map.
    - df (pd.DataFrame): The dataset containing player, map, and agent information.

    Returns:
    - str: The most likely agent the player will play on the given map.
    """
    # Filter the dataset for the given player and map
    filtered_df = df[(df['Name'] == player) & (df['Map'] == map_name)]
    
    if filtered_df.empty:
        return f"No data available for player '{player}' on map '{map_name}'."
    
    # Find the agent with the highest frequency
    most_likely_agent = (
        filtered_df['Agent']
        .value_counts()
        .idxmax()
    )
    
    return most_likely_agent

# Example usage
player_name = "invy"
map_name = "Pearl"
predicted_agent = predict_agent(player_name, map_name, df)
print(f"The most likely agent for {player_name} on {map_name} is {predicted_agent}.")
```
