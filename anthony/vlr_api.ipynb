```python
import pandas as pd
import json
import re
```

```python
import requests

# Base URL
url = "https://vlr.orlandomm.net/api/v1/results"

# Optional: Set query parameters
params = {
    "page": 2,
    "limit": 200,
}

# Make the GET request
response = requests.get(url, params=params)

data = response.json()
```

```python
data
```

```python
df = pd.DataFrame(data)
matches = pd.json_normalize(df["data"])
```

```python
matches["team1"] = pd.json_normalize(matches["teams"])[0]
matches["team2"] = pd.json_normalize(matches["teams"])[1]
```

```python
matches.drop(columns=["teams"], inplace=True)
```

```python
matches
```

```python
challengers_matches = matches[matches["tournament"].str.contains(r'\bchallengers\b', case=False, na=False, regex=True)]
vct_matches = matches[matches["tournament"].str.contains("Champions", case=False, na=False, regex=True)]
```
